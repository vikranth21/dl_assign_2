```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1,1,0.001)
h11 = 1/(1+np.exp(-(500*x+30)))
h12 = 1/(1+np.exp(-(500*x-30)))
h21 = h11 - h12
```

```python
plt.ylabel("h11(x)")
plt.xlabel("x")
plt.plot(x,h11)

plt.show()
```

```python
plt.ylabel("h12(x)")
plt.xlabel("x")
plt.plot(x,h12)
plt.show()
```

```python
plt.ylabel("h21(x)")
plt.xlabel("x")
plt.plot(x,h21)
plt.show()
```
